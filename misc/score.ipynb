```python
import operator
import urllib.request, json

url = "https://gist.githubusercontent.com/m-malikov/352e6c0ffa222e25f30d223419b92f89/raw/f3138771ddaf128e5d91817be5618decde8d64f7/tracker_data.json"
data = {}

with urllib.request.urlopen(url) as file:
    data = json.loads(file.read().decode())

print("Общее число сайтов в выборке: ", len(data))
```

```python
data['duckduckgo.com']
```

```python
print("Сайты, которые используют трекеры в куки: ")

trackersCookie = {}

for site in data:
    count = 0
    for cookie in data[site]["cookies"]:
        if cookie["is_thirdparty"]:
            if cookie["is_tracker"]:
                count += 1
    trackersCookie.update({site: count})
```

```python
print("Сайты, которые используют fingerprinting:\n")

fingerprinting = set()

for site in data:
    if 'fingerprinting' in data[site]:
        if data[site]['fingerprinting']['canvas']['is_fingerprinting']:  
            fingerprinting.update({site})
```

```python
print("Сайты, которые используют не анонимные гугл-запросы: ")

googleAnalytics = dict()
for site in data:
    if 'trackers' in data[site]['google_analytics']:
        count = 0
        for obj in data[site]['google_analytics']['trackers']:
            if not obj['anonymize_ip']:
                count += 1
        googleAnalytics.update({site: count})
```

```python
results = dict()
for site in data:
    count = 0.1*trackersCookie[site]
    if site in googleAnalytics.keys():
         count += 1*googleAnalytics[site]
    if site in fingerprinting:
        count += 10
    results.update({site: -1*count})

sorted(results.items(), key=lambda x: x[1])
```

```python
with open('score.json', 'w') as fp:
    json.dump(results, fp)
```
